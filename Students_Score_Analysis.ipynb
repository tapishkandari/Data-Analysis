# Analysis Of Students Scores Data

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head(5))
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop Unnamed column 

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df)
```

```python
df.head()
```

# Gender Distribution 

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
## from the above chart we have analysed that the no. of females in the given data is more than males.
```

# Parent education Impact on Students Scores

```python
gp = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gp, annot = True)
plt.title("Relationship between parents education and students score")
plt.show()
```

```python
#from the above chart we concluded that higher education of parents have a good impact on children scores.
```

# Parent Marital Status impact on Stdudents Scores 

```python
gp1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gp1, annot = True)
plt.title("Relationship between parents marital status and students score")
plt.show()
```

```python
#from the above chart Parent marital status have negligible effect.
```

# Distribution of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()
my_list = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

l = ["group A","group B","group C","group D","group E"]

plt.title("Distributions of ethnic groups")
plt.pie(my_list, labels = l, autopct ="%1.2f%%")
plt.show()
```

```python
#from above chart it is clear that group C Ethnicity is largest among others.
```
